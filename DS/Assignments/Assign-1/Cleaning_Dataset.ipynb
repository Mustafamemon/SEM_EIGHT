### 1. IMPORT

```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from nltk.metrics import edit_distance

print('Successfully Import')
```

### 2.Dataset and Base directory path¶


```python
DATASET_PATH = os.path.join(os.path.abspath(""), "data")

DATASET_1 = os.path.join(DATASET_PATH, "data.json")
DATASET_2 = os.path.join(DATASET_PATH, "data1.csv")
DATASET_3 = os.path.join(DATASET_PATH, "data2.csv")
DATASET_4 = os.path.join(DATASET_PATH, "data3.csv")
DATASET_5 = os.path.join(DATASET_PATH, "hcvdat0.csv")
```

### 3. Loading dataset into pandas frames¶


```python
df1 = pd.read_json(DATASET_1)
print('DF-1 :\n',df1.head(),'\n')

df2 = pd.read_csv(DATASET_2,index_col=0)
print('DF-2 :\n',df2.head(),'\n')

df3 = pd.read_csv(DATASET_3,index_col=1)
print('DF-3 :\n',df3.head(),'\n')

df4 = pd.read_csv(DATASET_4,index_col=0)
print('DF-4 :\n',df4.head(),'\n')

df5 = pd.read_csv(DATASET_5,index_col=0)
print('DF-5 :')
df5.head()
```

### 4. Check null / missing values in Dataset

```python
print(df1.isnull().any(),'\n')

print(df2.isnull().any(),'\n')

print(df3.isnull().any(),'\n')

print(df4.isnull().any(),'\n')

print(df5.isnull().any(),'\n')
```

### 5. Fill missing or nan with MEAN

```python
df2.fillna(df2.mean(),inplace=True)
print(df2)

df5.fillna(df5.mean(),inplace=True)
df5.head()
```

### 6. Check & Remove Outlier

```python
Q1 = df5.quantile(0.25)
Q3 = df5.quantile(0.75)
IQR = Q3 - Q1
detect_outlier = (df5 < (Q1 - 1.5 * IQR)) | (df5 > (Q3 + 1.5 * IQR))
detect_outlier.head()
print('BEFORE DROPPING OUTLIER ROWS : ',end='\n')

print(df5.shape)
df5 = df5[~((df5 < (Q1 - 1.5 * IQR)) |(df5 > (Q3 + 1.5 * IQR))).any(axis=1)]
print('AFTER DROPPING OUTLIER ROWS : ',end='\n')
print(df5.shape)
```

### 7. Check & Remove Inconsistency

```python
print(df5['Sex'].unique())
df5['Sex'] = df5['Sex'].str.lower()

print(df5['Sex'].unique())

df5['Sex_m'] = df5['Sex'].map(lambda x: edit_distance(x, 'm'))
df5['Sex_f'] = df5['Sex'].map(lambda x: edit_distance(x, 'f'))

df5.loc[ df5['Sex_m'] >= 2, 'Sex'] = 'm'

df5.loc[ df5['Sex_f'] >= 2, 'Sex'] = 'f'

print(df5['Sex'].unique())
df5.drop(['Sex_m','Sex_f'],axis=1)
```
